```python
# Lets see total runs in each inning

total_score_df = delivery.groupby(["match_id", "inning"]).sum()['total_runs'].reset_index()
total_score_df
```

```python
# Now, we will consider only the first inning of each match.
total_score_df1 = total_score_df[total_score_df['inning'] == 1]
total_score_df1
```

#### Merge matches dataframe with total_runs_df1 dataframe to get first innings scores in matches dataframe, we will leave out innings column as it is 1st innings always

```python
match_df = matches.merge(total_score_df1[['match_id','total_runs']], left_on = 'id', right_on= 'match_id')
match_df.head()
```

```python
match_df.shape
```

Number of teams 

```python
match_df.team1.unique()
```

```python
# We can see there are some teams like 'Rising Pune Supergiant', 'Gujarat Lions' who are not playing ipl 

teams = [
    'Royal Challengers Bengaluru',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Sunrisers Hyderabad',
    'Delhi Capitals', 
    'Punjab Kings','Lucknow Super Giants', 'Gujarat Titans']
```

```python
# 'Sunrisers Hyderabad' and 'Deccan Chargers' both are same teams, DD and DC, PBKS and KXIP are also the same 

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Punjab Kings')

match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
```

Filtering the dataframe, by keeping the matches which are relevant for current prediction (remove matches of teams who are not playing ipl anymore)

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df['team1'].unique()
```

```python
# Duckworth-Lewis-Stern (DLS) method
match_df['method'].value_counts()
```

15 IPL matches had results influenced by the DLS method. We will not consider matches where DLS is applied

```python
match_df = match_df[match_df['method'] != 'D/L']
match_df['method'].value_counts()
```

```python
# extracting features from match dataset
match_df = match_df[['match_id','city','winner','total_runs']]
match_df.head()
```

```python
# merging

delivery_df = match_df.merge(delivery, on = 'match_id')
delivery_df.head()
```

```python
# Now, let's consider only the second inning of each match.
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.head()
```

```python
delivery_df.shape
```

```python
# Lets now get current runs 
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
delivery_df.head()
```

- We can see that after each ball, the number of runs scored is being counted.

```python
# now we will find required runs
#required runs = total runs - current score
delivery_df['required_runs'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
# Now lets see how many balls left 

delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
delivery_df.head()
```

```python
# now lets see how many wickets are left
# If a player is out, it is recorded next to their name; otherwise, it is marked as NaN.
# We will fill nan with 0

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df.head()
```

```python
# Wherever there is a name, we will replace it with 1.
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x == '0' else "1")

#Here,`x if x == 0` ensures that if the value is 0,it remains unchanged,and it replaces other values (names) with 1

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
# lets count how many wickets have been taken

wickets= delivery_df.groupby("match_id")['player_dismissed'].cumsum().values
wickets
```

```python
# Add wickets_left column to the delivery_df dataframe

delivery_df['wickets_left'] = 10- wickets
delivery_df.head()
```

```python
delivery_df.shape
```

```python
# Current run rate (crr)= runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# required run rate:(rrr) 
delivery_df['rrr'] = (delivery_df['required_runs']*6)/delivery_df['balls_left']
```

```python
# If winner and batting team are same then result will be 1 esle 0
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

```python
delivery_df['result'].value_counts()
```

```python
# Now lets extract the columns that are required for our predictions
final_df = delivery_df[['batting_team','bowling_team','city','required_runs','balls_left','wickets_left','total_runs_x',
            'crr','rrr','result']]
final_df = final_df.sample(final_df.shape[0])
final_df.head()
```

```python
final_df.describe()
```

```python
final_df['crr'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(subset = ['crr'], inplace = True)
```

```python
# In CRR, we have `inf` values. 

final_df = final_df[final_df['balls_left'] !=0]
final_df.describe()
```

```python
final_df.isnull().sum()
```

```python
# THere are null values so we will drop na values
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

## Selecting Independent and Dependent Features

```python
X = final_df.iloc[:,:-1]
X.head()
```

```python
y = final_df['result']
```

#### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 40)
```

#### Encoding

```python
# Column transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('categorical', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city']) 
], remainder='passthrough')
```

Transform the data and fit a Logistic Regression Model 

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ('transform', trf),
    ('model',LogisticRegression(solver = 'liblinear'))
])
```

We are using LogReg as we can get probability of winning for each team 

```python
pipe.fit(X_train, y_train)
```

```python
# Predicting
y_pred = pipe.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
pipe.predict_proba(X_test)   # probability of winning of each team  
```
